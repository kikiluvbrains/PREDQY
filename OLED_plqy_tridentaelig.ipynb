```python
from rdkit import Chem
from rdkit.Chem import AllChem

# Function to generate Morgan fingerprints
def generate_morgan_fingerprint(smiles, radius=2, num_bits=2048):
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        fingerprint = AllChem.GetMorganFingerprintAsBitVect(mol, radius, nBits=num_bits)
        return fingerprint
    else:
        return None

# Keep track of indices with faulty SMILES strings
faulty_indices = []

# Generate Morgan fingerprints for the SMILES strings
filtered_solvent_smiles_fingerprints = []

for i, smiles in enumerate(solvent_smiles):
    fingerprint = generate_morgan_fingerprint(smiles)
    if fingerprint is not None:
        filtered_solvent_smiles_fingerprints.append(fingerprint)
    else:
        # Skip faulty SMILES and record the index
        faulty_indices.append(i)

# Print the content of filtered_structures_fingerprints in the specified format
print("Filtered Structures:")
for i, smiles in enumerate(solvent_smiles):
    print(f"Entry {i + 1}: {smiles}")

# Note: Printing the fingerprints directly might not be very informative since they're binary vectors
# If you want to see them as bit strings or another format, you'd need to convert them accordingly.

# Print indices with faulty SMILES
print("Faulty Indices:", faulty_indices)
```

```python
# Assuming filtered_chromophore_fingerprints and filtered_solvent_fingerprints
# contain fingerprint bit strings of the same length and are aligned

combined_fingerprints = []

for chromophore_fp, solvent_fp in zip(filtered_chromophore_fingerprints, filtered_solvent_smiles_fingerprints):
    # For bit strings, concatenation can be done directly with the + operator
    combined_fp = chromophore_fp + solvent_fp
    combined_fingerprints.append(combined_fp)

# Now combined_fingerprints contains the concatenated fingerprints
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test, paper_train, paper_test = train_test_split(
    structures_fp_array, Quantum_yield_array, paper,
    test_size=0.2, random_state=42
)

# Reshape the input data for LSTM
X_train_reshaped = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_reshaped = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, BatchNormalization, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import LearningRateScheduler, EarlyStopping
from tensorflow.keras.metrics import MeanAbsoluteError
import torch
import torch.nn as nn
import torch.optim as optim
from torch.optim.lr_scheduler import LambdaLR
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, BatchNormalization, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from keras import backend as K


model = Sequential()
# Replacing Bidirectional LSTM with Unidirectional LSTM
model.add(LSTM(300, activation="relu", return_sequences=True, input_shape=(X_train_reshaped.shape[1], X_train_reshaped.shape[2])))
model.add(Dropout(0.4))  # Add dropout with a rate of 40%
model.add(LSTM(300, activation="relu", return_sequences=True))
model.add(Dropout(0.1))  # Add dropout with a rate of 10%
model.add(LSTM(200, activation="relu"))  # Fixed missing comma here
# Uncomment if using GlobalMaxPooling1D
# model.add(GlobalMaxPooling1D(data_format='channels_last'))

# Initialize the optimizer
optimizer = Adam(learning_rate=0.001)

# Assuming 'denselayersnodes' is defined elsewhere as the number of nodes in your final Dense layer
model.add(Dense(denselayersnodes, activation=None))  # Use 'sigmoid' for binary outcomes

# Compile the model
model.compile(loss='mse', optimizer=optimizer, metrics=['mae'])

# Define the learning rate scheduler
def lr_schedule(epoch):
    return 0.001 * (0.9 ** epoch)

lr_scheduler = LearningRateScheduler(lr_schedule)

# Define early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=300, restore_best_weights=True)

# Train the model
history = model.fit(X_train_reshaped, y_train, epochs=20000, batch_size=32, validation_split=0.2, callbacks=[lr_scheduler, early_stopping])
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_absolute_error

# Make predictions on the test set
y_pred = model.predict(X_test_reshaped)

# Scatter plot of actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5, s=10)  # Use a smaller marker size with s=10
plt.title('Actual vs. Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)  # Diagonal line

# Calculate R and MAE
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

# Add a text box with R and MAE values
textstr = f'$R^2={r2:.2f}$\n$MAE={mae:.2f}$'
# Position: right top corner
plt.text(0.95, 0.05, textstr, transform=plt.gca().transAxes, fontsize=14,
         verticalalignment='bottom', horizontalalignment='right',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))

plt.show()

# Display R and MAE in the console as well
print(f'R: {r2}')
print(f'Mean Absolute Error (MAE): {mae}')
```
